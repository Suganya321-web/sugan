```
import pandas as pd
import sqlite3
file_name = '/content/Data Analysis.csv'

df = pd.read_csv(file_name)
print("Data loaded. Shape:", df.shape)
df.head()
```

```
conn = sqlite3.connect(':memory:')
df.to_sql('data', conn, index=False, if_exists='replace')

def sql(query):
    return pd.read_sql_query(query, conn)

print("Table 'data' ready. Columns:", list(df.columns))
```

```

sql("""SELECT *, (Quantity * UnitPrice) AS TotalAmount FROM data WHERE (Quantity * UnitPrice) > 1000 LIMIT 10""")
sql("""SELECT CustomerID, COUNT(*) AS total_orders FROM data GROUP BY CustomerID ORDER BY total_orders DESC""")
sql("""SELECT Product, SUM(Quantity * UnitPrice) AS total_sales FROM data GROUP BY Product ORDER BY total_sales DESC""")

sql("""SELECT AVG(Quantity * UnitPrice) AS avg_order_value FROM data""")

sql("""SELECT *, (Quantity * UnitPrice) AS TotalAmount FROM data ORDER BY TotalAmount DESC LIMIT 5""")
```
